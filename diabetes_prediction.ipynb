```python
import sklearn
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
# loading the csv data to a Pandas DataFrame
diabetes_data = pd.read_csv('diabetes_data.csv')
```

```python
X = diabetes_data.drop('Outcome', axis=1)
y = diabetes_data['Outcome']
```

```python
print(y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
     

print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
log_model = LogisticRegression()
log_model.fit(X_train, Y_train)
log_pred = log_model.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(Y_test, log_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)
knn_pred = knn.predict(X_test)

print("KNN Accuracy:", accuracy_score(Y_test, knn_pred))
```

```python
svm_model = SVC()
svm_model.fit(X_train, Y_train)
svm_pred = svm_model.predict(X_test)

print("SVM Accuracy:", accuracy_score(Y_test, svm_pred))
```

```python
rf = RandomForestClassifier(n_estimators=150)
rf.fit(X_train, Y_train)
rf_pred = rf.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(Y_test, rf_pred))
```

```python
# Prediction Function

def predict_diabetes(input_data):
    # convert to numpy array
    input_array = np.asarray(input_data)

    # reshape for single sample
    reshaped = input_array.reshape(1, -1)

    # scale input
    scaled_input = scaler.transform(reshaped)

    # prediction
    pred = log_model.predict(scaled_input)

    if pred[0] == 1:
        return "The person is diabetic"
    else:
        return "The person is not diabetic"
```

```python
sample = [2, 130, 70, 20, 79, 27.1, 0.5, 30]  # sample values
predict_diabetes(sample)
```
